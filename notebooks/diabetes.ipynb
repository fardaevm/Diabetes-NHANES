```python
import pandas as pd
```

```python
diabetes_survey = pd.read_sas('../data/DIQ_L.xpt')
demographics = pd.read_sas('../data/demographics.xpt')
ghb = pd.read_sas('../data/GHB_L.xpt')
glucose = pd.read_sas('../data/GLU_L.xpt')
body_measure = pd.read_sas('../data/BMX_L.xpt')
```

```python
df = demographics.merge(diabetes_survey, on="SEQN", how="inner")
df = df.merge(body_measure, on="SEQN", how="inner")
df = df.merge(glucose, on="SEQN", how="inner")
df = df.merge(ghb, on="SEQN", how="left")
```

```python
df
```

```python
df.columns
```

```python
selected_cols = [
    # Unique ID
    
    # Diabetes status
    'DIQ010',     # Told by doctor: 1 = Yes, 2 = No, 3 = Borderline
    
    # Biomarkers
    'LBXGLU',     # Fasting glucose (mg/dL)
    'LBXGH',      # A1C (%)

    # 📏 Body measures
    'BMXBMI',     # BMI
    'BMXWT',      # Weight (kg)
    'BMXHT',      # Height (cm)
    'BMXWAIST',   # Waist circumference (cm)
    

    # 🧍 Demographics
    'RIDAGEYR',   # Age
    'RIAGENDR',   # Gender (1 = Male, 2 = Female)
    'RIDRETH3',   # Race/Ethnicity
    'DMDEDUC2',   # Education level
    'INDFMPIR',   # Income-to-poverty ratio
    'RIDEXPRG'    # Pregnancy status
]

df_clean = df[selected_cols].copy()
```

```python
df_clean.rename(columns={
    'DIQ010': 'diabetes',
    'LBXGLU': 'fasting_glucose',
    'LBXGH': 'a1c',
    'BMXBMI': 'bmi',
    'BMXWT': 'weight_kg',
    'BMXHT': 'height_cm',
    'BMXWAIST': 'waist_cm',
    'RIDAGEYR': 'age',
    'RIAGENDR': 'gender',
    'RIDRETH3': 'ethnicity',
    'DMDEDUC2': 'education',
    'INDFMPIR': 'poverty_index',
    'RIDEXPRG': 'pregnancy'
}, inplace=True)
```

```python
df_clean.head()
```

### Race Category
* 1 - Mexican American
* 2	- Other Hispanic
* 3 - Non-Hispanic White
* 4 - Non-Hispanic Black
* 6 - Non-Hispanic Asian
* 7	- Other race — including multi-racial
### Gender Category
* 1 - Male
* 2 - Female
### Poverty Index
* 0.00 – 1.00	At or below the poverty line
* 1.01 – 2.00	Just above poverty (low income)
* 2.01 – 3.00	Moderate income
* 3.01 - 4.99	Higher income
* 5.00 -  above average income
### Pregnancy
* 1 - pregnant
* 2 - not pregnant
* 3 - not sure, cannot ascertain

```python
# Dropping na values from fasting glucose and a1c
df_clean = df_clean.dropna(subset=['fasting_glucose', 'a1c'])
```

```python
# Dropping data on pregnant women
df_clean = df_clean[df_clean['pregnancy']!=1]

# Dropping pregnancy column
df_clean.drop(columns=['pregnancy'], inplace=True)
```
