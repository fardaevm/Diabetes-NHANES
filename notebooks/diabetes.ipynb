```python
import pandas as pd
```

```python
diabetes_survey = pd.read_sas('../data/external/DIQ_L.xpt')
demographics = pd.read_sas('../data/external/demographics.xpt')
ghb = pd.read_sas('../data/external/GHB_L.xpt')
glucose = pd.read_sas('../data/external/GLU_L.xpt')
body_measure = pd.read_sas('../data/external/BMX_L.xpt')
```

```python
df = demographics.merge(diabetes_survey, on="SEQN", how="inner")
df = df.merge(body_measure, on="SEQN", how="inner")
df = df.merge(glucose, on="SEQN", how="inner")
df = df.merge(ghb, on="SEQN", how="left")
```

```python
selected_cols = [
    
    # Biomarkers
    'LBXGH',      # A1C (%)

    # Body measures
    'BMXBMI',     # BMI
    'BMXWT',      # Weight (kg)
    'BMXHT',      # Height (cm)
    'BMXWAIST',   # Waist circumference (cm)
    

    # Demographics
    'RIDAGEYR',   # Age
    'RIAGENDR',   # Gender (1 = Male, 2 = Female)
    'RIDRETH3',   # Race/Ethnicity
    'INDFMPIR',   # Income-to-poverty ratio
    'RIDEXPRG'    # Pregnancy status
]

df_clean = df[selected_cols].copy()
```

```python
df_clean.rename(columns={
    'DIQ010': 'diabetes',
    'LBXGLU': 'fasting_glucose',
    'LBXGH': 'a1c',
    'BMXBMI': 'bmi',
    'BMXWT': 'weight_kg',
    'BMXHT': 'height_cm',
    'BMXWAIST': 'waist_cm',
    'RIDAGEYR': 'age',
    'RIAGENDR': 'gender',
    'RIDRETH3': 'ethnicity',
    'DMDEDUC2': 'education',
    'INDFMPIR': 'poverty_index',
    'RIDEXPRG': 'pregnancy'
}, inplace=True)
```

```python
# Dropping na values from fasting glucose and a1c
df_clean = df_clean.dropna(subset=['fasting_glucose', 'a1c'])
```

```python
# Dropping data on pregnant women
df_clean = df_clean[df_clean['pregnancy']!=1]

# Dropping pregnancy column
df_clean.drop(columns=['pregnancy'], inplace=True)
```

```python
# classified diabetes into 3 categories: Diagnosed, Undiagnosed, No Diabetes
def classify_diabetes(row):
    if row['diabetes'] == 1:
        return 'Diagnosed'
    elif row['diabetes'] != 1 and (
        (pd.notnull(row['fasting_glucose']) and row['fasting_glucose'] >= 126) or 
        (pd.notnull(row['a1c']) and row['a1c'] >= 6.5)
    ):
        return 'Undiagnosed'
    else:
        return 'No Diabetes'
    
df_clean['diabetes_status'] = df_clean.apply(classify_diabetes, axis=1)
        
```
